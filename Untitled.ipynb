## P-Value
P-value helps to decide whether to reject the null hypothesis (H0) when a statistical test is performed. The smaller the p-value, the more likely it is to reject the null hypothesis.

If the result of the test is a small p (≤ 0.05), this is strong evidence that the null hypothesis is invalid. A large p (> 0.05) means that the alternative hypothesis is weak, so the null hypothesis cannot be rejected.

So in this step we will eliminate features with a p-value of 0.05 

```python
#Create a copy of x beacuse we are going to operate on x and we don't want to change this data
xCopy = x

#I create a list to know which features to remove.
dropList = []

#Each time we remove a feature with a p value higher than 0.05, the p values in all features change, so we use a for loop here.
for i in range(x.shape[1]):
    
    #Creating model and p-value series.
    modelSm = sm.OLS(y, xCopy).fit()
    pValuesSeries = modelSm.pvalues
    #first of all I check if the p value is 0.05.
    if pValuesSeries.max()> 0.05:
        #finding max p value
        maxPValue = pValuesSeries.idxmax()

        #drop the max p value
        xCopy = xCopy.drop(columns=maxPValue)
        pValuesSeries = pValuesSeries.drop(maxPValue)

        
        #append the wrong feature to remove list.
        dropList.append(maxPValue)
    #End the loop if there is no p value higher than 0.05 in the series.
    else:
        break

print(dropList)
```

## Feature Selection
I know which features to take out. Let's do it.

```python
# Appending BloodPressure I mentioned above that the correlation rate is low (skinthickness is already on the list so I'm not adding it).
dropList.append('BloodPressure')
#dropList.remove("Age")
# Drop wrong features.
x = x.drop(dropList, axis=1)
x
```

# TRAINING MODEL


## Train Test Split With K-Fold


```python
# Sets up 5-fold cross-validation, dividing the data into training and testing sets for model evaluation.
kf = KFold(n_splits=5, shuffle=True, random_state=42)
for train_index, test_index in kf.split(x):
    xTrain, xTest = x.iloc[train_index], x.iloc[test_index]
    yTrain, yTest = y.iloc[train_index], y.iloc[test_index]


# xTrain, xTest, yTrain, yTest = train_test_split( x, y, test_size=0.33, random_state=42)
xTrain.head()
```

```python
# Creating scale model
standartScaler = StandardScaler()

# Creating classifiers
xgboostClassifier = xgboost.XGBClassifier() 
catBoostClassifier = catboost.CatBoostClassifier() 
lightgbmClassifier = lightgbm.LGBMClassifier() 

#  Define machine learning pipelines, each consisting of two steps: data scaling using 'standardScaler' and applying a different classifier.
pipeXgboost = Pipeline(steps=[('scaler', standartScaler), ('xgboostModel', xgboostClassifier)])
pipeCatBoost = Pipeline(steps=[('scaler', standartScaler), ('catboostModel', catBoostClassifier)])
pipeLightgbm = Pipeline(steps=[('scaler', standartScaler), ('lightgbmModel', lightgbmClassifier)])

# A pipelist to random search
pipeList = [pipeXgboost,pipeCatBoost,pipeLightgbm]

# quick test
pipeXgboost.fit(xTrain,yTrain)
predXGB = pipeXgboost.predict(xTest)
cm = confusion_matrix(yTest,predXGB)
print(cm)
print(accuracy_score(yTest,predXGB))
```

## Creating Grids


```python

xgboostGrid = { 
    # Learning rate shrinks the weights to make the boosting process more conservative
    "xgboostModel__n_estimators": [100,200,300,400,500] ,
     
    # Learning rate shrinks the weights to make the boosting process more conservative
    "xgboostModel__learning_rate": [0.01, 0.1, 1] ,
    # Maximum depth of the tree, increasing it increases the model complexity.
    "xgboostModel__max_depth": range(5,17,4),
    # Gamma specifies the minimum loss reduction required to make a split.
    "xgboostModel__subsample": [0.5,1,3,5],
    # Percentage of columns to be randomly samples for each tree.
    "xgboostModel__min_child_weight": [1,3,5,7],

   }

catboostGrid = {
    # "n_estimators" determines the number of trees used in the CatBoost model.
    'catboostModel__n_estimators' : [100,200,300],

    #  "max_depth" determines the maximum depth of the trees. 
    'catboostModel__max_depth' : [3,4,5]


}

lightgbmGrid = {
    # "n_estimators" represents the number of boosting rounds or trees in the LightGBM model.
    'lightgbmModel__n_estimators' : [100,200,300],

    # "Learning_rate" controls the step size at each iteration while updating the model.
    'lightgbmModel__learning_rate' : [0.01,0.3],

    # "Max_depth" determines the maximum depth of the individual decision trees in the LightGBM model.
    'lightgbmModel__max_depth':[3,12]
}

# Create a grid list to random search
gridList = [xgboostGrid,catboostGrid,lightgbmGrid]
# Set up score
scoring = ['recall']
# Set up the k-fold cross-validation
kFold = StratifiedKFold(n_splits=3, shuffle=True, random_state=0)
```

## Fitting Model

```python
# Define random search
randomResultList = []
predictList = []
for i in range(len(pipeList)):

    randomSearch = RandomizedSearchCV(pipeList[i], 
                            gridList[i], 
                            n_iter=48,
                            scoring=scoring, 
                            refit='recall', 
                            n_jobs=-1, 
                            cv=kFold, 
                            verbose=0,
                            pre_dispatch = 2)
    # Fit random search
    randomResultList.append(randomSearch.fit(xTrain, yTrain)) 
    
    # Predict random search
    predictList.append(randomSearch.predict(xTest))
```

## Scoring

```python
# Create a table for scoring 
modelList = ['xgboost','catboost','lightgbm']
for i in range(len(randomResultList)):
    
    
    print(
        f'''
        {'='*50}{modelList[i]}{'='*50}

        The Accuracy score: {accuracy_score(yTest,predictList[i])}
        The Confusion Matrix:
        
        {confusion_matrix(yTest,predictList[i])}
        The best score:{randomResultList[i].best_score_:.4f}
        The best crore standart deviation: {round(randomResultList[i].cv_results_['std_test_recall'][randomResultList[i].best_index_], 4)}
        The best hyperparameters are: {randomResultList[i].best_params_}
        ''')

    
```

## Saving Model

```python
# The best model is xgboost. Let's save the model
fileName = "model.pkl"
bestModel = randomResultList[0]

# Saving Model
pickle.dump(bestModel,open(fileName,'wb'))
xgboostBestModel = pickle.load(open(fileName,'rb'))

# Testing
print(xgboostBestModel.predict(xTest))
```
